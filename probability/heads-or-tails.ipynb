```python
import random
%matplotlib inline
import matplotlib.pyplot as plt
import numpy

plt.rcParams["figure.figsize"] = (20,10)
```

```python
flips = []
number_of_flips = 10000

for i in xrange(0, number_of_flips):
    flips.append(random.randint(0,1))
```

```python
flips[0:10]
```

```python
percent_heads = []
for j in xrange(1, number_of_flips + 1):
    percent_heads.append(numpy.mean(flips[0:j]))
```

```python
percent_heads[0:10]
```

```python
plt.plot(xrange(1,number_of_flips+1), percent_heads, color='blue')
plt.plot(xrange(0,number_of_flips+1), [.5 for number in xrange(number_of_flips+1)], color='0.75')

plt.ylim(0, 1.0)
plt.title("Percentage of Heads with {} Coin Flips".format(number_of_flips), fontsize=24)
plt.xlabel("Number of Flips", fontsize=18)
plt.ylabel("Percent Heads", fontsize=18)

plt.show()
```
